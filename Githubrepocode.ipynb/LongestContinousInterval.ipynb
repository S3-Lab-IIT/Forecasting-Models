# Longest Continous Interval 

### Importing Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Data Collection

##### Data File: 06 Feb 2024
##### Time interval: 2024-02-06 02:34:55.870 : 2024-02-06 23:59:59.420

```python
bsp_data = pd.read_pickle(r"C:\Users\91780\Desktop\IIT bhilai internship\Code\Githubrepocode.ipynb\Dataset\clean_06_02_24.pkl")
region_wise = pd.read_pickle(r"C:\Users\91780\Desktop\IIT bhilai internship\Code\Githubrepocode.ipynb\Dataset\regionwise_columns.pickle")
df = bsp_data[region_wise['Stand_13-18']]
```

```python
# running a loop to calculate the difference between timestamps
for idx in range(1,len(df.index)):
    prev=df.index[idx-1]
    curr=df.index[idx]
    df.loc[df.index[idx],'diff']=(curr-prev).total_seconds()
```

### Visualization of the difference calculated

```python
import plotly.graph_objects as go

# Creating a line plot using Plotly
fig = go.Figure()

fig.add_trace(go.Scatter(x=df.index,
                             y=df['diff']/60,
                             mode='lines',
                             name="score"))

# Customizing the layout
fig.update_layout(
    title=f'Time Series Forecast with ARIMA',
    xaxis_title='time',
    yaxis_title='[9:13]',
    showlegend=True,
)

# Display the plot
fig.show()
```

```python
# Initialize variables to track the longest sequence of 1s
max_length = 0
max_start = None
max_end = None

current_length = 0
current_start = None

for time, value in df['diff'].items():
    if value == 1:
        if current_length == 0:
            current_start = time
        current_length += 1
    else:
        if current_length > max_length:
            max_length = current_length
            max_start = current_start
            max_end = time - pd.Timedelta(seconds=1)
        current_length = 0

# Check the last sequence if it was the longest and ended at the end of the DataFrame
if current_length > max_length:
    max_length = current_length
    max_start = current_start
    max_end = time

print("\nLongest interval of continuous 1s:")
print(f"Start time: {max_start}")
print(f"End time: {max_end}")
print(f"Length: {max_length} seconds")
```

```python
df = df.loc[max_start:max_end]
```
